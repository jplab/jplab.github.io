```python
list_cites = []
labels = []
for software in cas:
    name = software['name']
    list_cites += [cas_p[name]]
    labels += [software['name']+ ' Primary 52']
colors = distinctipy.get_colors(len(list_cites))
```

```python
plt.stackplot(range(2000,2022),
              list_cites,
              labels=labels,
              colors = colors,
              alpha=0.8)

plt.legend(loc=2, fontsize='large')
plt.show()
```
